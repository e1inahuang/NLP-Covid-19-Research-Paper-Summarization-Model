```python
df = df.sort_values(by='publish_time')
df
```

```python
def parse_dates(date):
    if '-' in date:
        return pd.to_datetime(date, format='%Y-%m-%d', errors='coerce')
    else:
        return pd.to_datetime(date, format='%Y', errors='coerce')

df['publish_time'] = df['publish_time'].apply(parse_dates)
```

```python
count = df[df['publish_time'].dt.year > 2001].shape[0]
count
```

```python
df['publish_time'] = df['publish_time'].dt.year
df = df[(df['publish_time'] >= 2019) & (df['publish_time'] <= 2022)]
```

```python
df
```

```python
import pandas as pd
from langdetect import detect, DetectorFactory
from tqdm import tqdm

# This is to ensure reproducibility of langdetect
DetectorFactory.seed = 0

# Function to detect language
def detect_language(text):
    try:
        return detect(text)
    except:
        # If detection fails, return 'unknown'
        return 'unknown'

# tqdm.pandas() will allow us to use progress_apply instead of apply
tqdm.pandas(desc="Detecting Language")

# Apply the function with a progress bar
df['language'] = df['title'].progress_apply(detect_language)

# Filter out non-English titles
df_english_only = df[df['language'] == 'en']

print(df_english_only)
```
